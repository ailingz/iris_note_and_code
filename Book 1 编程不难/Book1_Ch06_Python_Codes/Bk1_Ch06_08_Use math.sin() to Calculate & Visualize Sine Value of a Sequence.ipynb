### 利用 `math.sin()` 计算数列列表正弦值并可视化

```python
import math
import matplotlib.pyplot as plt
```

```python
print(dir(math))
```

```python
# 计算正弦值
x_start = 0 # 弧度值
print(math.sin(x_start))
```

```python
x_end = 2 * math.pi # 弧度值
print(math.sin(x_end))
```

```python
print(math.sin(0))           # 0
print(math.sin(math.pi/4))   # 45
print(math.sin(math.pi/2))   # 90
print(math.sin(math.pi))     # 180
print(math.sin(math.pi * 2)) # 360
```

```python
# 等差数列 arithmetic progression: a_n = a_1 + d(n – 1)
# 数列元素数量
num = 37

# 计算公差
step = (x_end - x_start) / (num - 1)

# 利用列表生成式 list comprehension 生成等差数列列表
x_array = [x_start + i * step for i in range(num)]
print(x_array)
```

```python
# 生成等长全0列表，等价于 zero_array = [0] * len(x_array)
zero_array = [0 for i in range(num)]
```

```python
# 可视化等差数列
plt.plot(x_array, zero_array,
         marker='.',
         markersize=8,
         markerfacecolor='w',
         markeredgecolor='k')
plt.text(x_start-0.1, 0.005, '0')
plt.text(x_end-0.1, 0.005, r'$2\pi$') # r 表示原始字符串 raw string
plt.axis('off') # 关闭坐标轴的显示
plt.show()
```

```python
# 正弦 sin(x) 列表
y_array = [math.sin(x_idx) for x_idx in x_array]
print(y_array)
```

```python
# 可视化正弦函数
plt.plot(x_array, y_array,
         marker='.',
         markersize=8,
         markerfacecolor='w',
         markeredgecolor='k')
plt.text(x_start-0.1, -0.1, '0')
plt.text(x_end-0.1, 0.1, r'$2\pi$')
plt.axhline(y=0, c='k', ls='--', lw=1) # color:c, linestyle:ls, linewidth:lw
plt.axis('off')
plt.show()
```
