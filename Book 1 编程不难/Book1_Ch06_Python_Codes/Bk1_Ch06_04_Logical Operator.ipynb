### 逻辑运算符

- 和运算符 `and` 当左右两边的操作数都为 `True` 时，返回 `True`，否则返回 `False`
- 或运算符 `or` 当左右两边的操作数至少有一个为 `True` 时，返回 `True`，否则返回 `False`
- 取非运算符 `not` 对一个布尔类型的操作数取反，如果操作数为 `True`，返回 `False`；否则返回 `True`



```python
# 和 and
print(True and True)
print(True and False)
print(False and True)
print(False and False)
```

```python
print(True and 1)
print(1 and True)
print(1 and False)
print(0 and True)
print(False and 0)
print(0 and False)
```

```python
# 或 or
print(True or True)
print(True or False)
print(False or True)
print(False or False)
```

```python
print(True or 1)
print(1 or True)
print(1 or False)
print(0 or True)
print(False or 0)
print(0 or False)
```

```python
# 非 not
print(not True)
print(not False)
```
