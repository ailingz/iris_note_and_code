### 线性回归

```python
# 导入包
import random
import statistics
import matplotlib.pyplot as plt
```

```python
# 产生数据
num = 50
random.seed(0)

# 自变量: 在区间[0, 10)内均匀分布的50个随机数
x_data = [random.uniform(0, 10) for _ in range(num)]

# 噪音
noise = [random.gauss(0, 1) for _ in range(num)]

# 因变量: 0.5*x + 1 + noise。0.5为斜率，1为截距
y_data = [0.5 * x_data[idx] + 1 + noise[idx] for idx in range(num)]
```

```python
# 绘制散点图
fig, ax = plt.subplots()

ax.scatter(x_data, y_data)

ax.set_xlabel('x'); ax.set_ylabel('y')
ax.set_aspect('equal', adjustable='box')
ax.set_xlim(0, 10); ax.set_ylim(-2, 8)
ax.grid()
```

```python
# 一元线性回归
# 调用 statistics.linear_regression() 函数计算一元线性回归模型的斜率和截距
slope, intercept = statistics.linear_regression(x_data, y_data)
```

```python
# 生成一个等差数列
start, end, step = 0, 10, 0.5

x_array = []
x_i = start

while x_i <= end:
    x_array.append(x_i)
    x_i += step
```

```python
# 计算 x_array 预测值
y_array_predicted = [slope * x_i + intercept for x_i in x_array]
```

```python
# 可视化一元线性回归直线
fig, ax = plt.subplots()

ax.scatter(x_data, y_data)
ax.plot(x_array, y_array_predicted, color='r')

ax.set_xlabel('x'); ax.set_ylabel('y')
ax.set_aspect('equal', adjustable='box')
ax.set_xlim(0, 10); ax.set_ylim(-2, 8)
ax.grid()
```
