```python
import math
import statistics
import random
import matplotlib.pyplot as plt
```

#### Q2. 利用 `math` 库计算半径为2的圆的周长和面积。

```python
r = 2
c = 2 * math.pi * r
a = math.pi * math.pow(r, 2)

print(f'Circumference is {c}')
print(f'Area is {a}')
```

#### Q3. 利用 `math` 库绘制 $f(x) = exp(x)$ 函数图像，$x$ 的取值范围为[−3, 3]。

```python
x_data = [i/10 for i in range(-30, 31)]
y_data = [math.exp(x) for x in x_data]

plt.plot(x_data, y_data)

plt.axhline(1, color='r')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x) = exp(x)$')
plt.grid()

plt.show()
```

#### Q4. 利用 `math` 库计算并绘制高斯函数 $f(x) = exp(-x^2)$ 图像，$x$ 的取值范围为[−3, 3]。

```python
x_data = [i/10 for i in range(-30, 31)]
y_data = [math.exp(-math.pow(x, 2)) for x in x_data]

plt.plot(x_data, y_data)

plt.xlabel(r'$x$')
plt.ylabel(r'$f(x) = exp(-x^2)$')
plt.grid()

plt.show()
```

#### Q5. 利用 `random` 和 `statistics` 库函数完成如下实验。抛一枚质地均匀六面色子1000次，色子点数为[1, 2, 3, 4, 5, 6]。可视化均值随抛掷次数变化。

```python
num = 1000
random.seed(0)

dices = [random.randint(1, 6) for _ in range(num)]
avg = [statistics.mean(dices[:i+1]) for i in range(num)]

plt.plot(range(1, num+1), avg)

plt.axhline(3.5, color='r')
plt.grid()

plt.show()
```

#### Q6. 利用 `random` 和 `statistics` 库函数完成如下实验。抛一枚质地不均匀色子1000次。点数[1, 2, 3, 4, 5, 6] 对应的概率分别为[0.2, 0.16, 0.16, 0.16, 0.16, 0.16]。
##### 可视化点数均值随抛掷次数变化。

```python
num = 1000
random.seed(300)

dices = [random.choices([1, 2, 3, 4, 5, 6], [0.2, 0.16, 0.16, 0.16, 0.16, 0.16])[0] for _ in range(num)]
avg = avg = [statistics.mean(dices[:i+1]) for i in range(num)]

plt.plot(range(1, num+1), avg)

plt.axhline(3.4, color='r')
plt.grid()

plt.show()
```
