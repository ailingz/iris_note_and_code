### 绘制箭头图

箭头图 quiver plot/vector plot

```python
import matplotlib.pyplot as plt
```

```python
# 定义二维列表
A = [[0, 5],
     [3, 4],
     [5, 0]]
```

```python
# 平面箭头图
# 自定义可视化函数
def draw_vector(vector, RGB):        # 自定义函数，用来绘制二维平面箭头
    plt.quiver(0, 0,                 # 参数0和0是箭头的起点坐标，这表示箭头起点在原点(0, 0)处
               vector[0], vector[1], # 参数 vector[0] 为箭头在 x 轴上的分量，vector[1] 是箭头在 y 轴上的分量
               angles='xy',          # 参数 angles='xy' 指定箭头应该以 x 和 y 轴的角度来表示
               scale_units='xy',     # 参数 scale_units='xy' 指定箭头的比例应该根据 x 和 y 轴的单位来缩放
               scale=1,              # 参数 scale=1 指定箭头的长度应该乘以的比例因子
               color=RGB,            # 参数 color=RGB 指定箭头的颜色
               zorder=1e5)           # 参数 zorder 用来指定艺术家图层序号. Artists with lower zorder values are drawn first
```

```python
fig, ax = plt.subplots()

# 第一行向量
v1 = A[0]
draw_vector(v1, '#FFC000') # Yellow

# 第二行向量
v2 = A[1]
draw_vector(v2, '#00CC00') # Green

# 第三行向量
v3 = A[2]
draw_vector(v3, '#33A8FF') # Blue

ax.axvline(x=0, c='k')
ax.axhline(y=0, c='k')

ax.set_xlabel('x1')
ax.set_ylabel('x2')

ax.grid()

ax.set_aspect('equal', adjustable='box')

ax.set_xbound(lower=-0.5, upper=5)
ax.set_ybound(lower=-0.5, upper=5)
```

```python
# 三维箭头图
# 自定义可视化函数
def draw_vector_3D(vector, RGB):                # 自定义函数绘制三维箭头图
    plt.quiver(0, 0, 0,                         # 参数0, 0, 0是箭头的起点坐标，即三维空间中的原点(0, 0, 0)处
               vector[0], vector[1], vector[2], # 参数 vector[0]、vector[1]、vector[2] 是箭头在 x、y、z 轴上的分量
               arrow_length_ratio=0,            # 参数 arrow_length_ratio=0 指定箭头的长度应该乘以的比例因子
                                                # 0表示箭头的长度将被设置为一个合适的长度，不乘以比例因子
               color=RGB,
               zorder=1e5)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d', proj_type='ortho')

# 第一列向量
v_1 = [row[0] for row in A]
draw_vector_3D(v_1, '#FF6600')

# 第二列向量
v_2 = [row[1] for row in A]
draw_vector_3D(v_2, '#FFBBFF')

ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.set_zlim(0, 5)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')

ax.view_init(elev=25, azim=30)
ax.set_box_aspect([1, 1, 1])
```
