### 用 `Seaborn` 生成热图

```python
# 导入包
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# 创建二维数据
data = np.random.rand(10, 10)

# 绘制热图
sns.heatmap(data, vmin=0, vmax=1,
            cmap='viridis',
            annot=True,
            xticklabels=True,
            yticklabels=True)
```

```python
# 从 seaborn 中导入鸢尾花样本数据
iris_sns = sns.load_dataset('iris')
```

```python
# 绘制热图
fig, ax = plt.subplots()

sns.heatmap(data=iris_sns.iloc[:, 0:-1],
            vmin=0, vmax=8,
            ax=ax,
            xticklabels=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'],
            yticklabels=False,
            cmap='RdYlBu_r')
```
