### 用 `Matplotlib` 生成平面等高线

```python
# 导入包
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 网格数据
x1 = np.arange(5)
x2 = np.arange(5)
xx1, xx2 = np.meshgrid(x1, x2)

print(xx1)
print(xx2)

for row in [[(xx1[i][j], xx2[i][j]) for i in range(5)] for j in range(5)]:
    print(row)
```

```python
# 创建二维数据
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# 绘制等高线图
plt.contour(X, Y, Z, levels=np.linspace(0, 8, 16+1), cmap='RdYlBu_r')

# 添加颜色图例
plt.colorbar()

plt.show()
```

```python
# 生成数据
x1_array = np.linspace(-3, 3, 121)
x2_array = np.linspace(-3, 3, 121)

xx1, xx2 = np.meshgrid(x1_array, x2_array)
ff = xx1 * np.exp(-xx1**2 - xx2**2)

# 等高线
fig, ax = plt.subplots()

CS = ax.contour(xx1, xx2, ff, levels=20, cmap='RdYlBu_r', linewidths=1)

fig.colorbar(CS)
ax.set_title('$f(x_1, x_2) = x_1 \exp(-x^2_1-x^2_2)$', fontsize=10)
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlim(xx1.min(), xx1.max())
ax.set_ylim(xx2.min(), xx2.max())
ax.grid(False)
ax.set_aspect('equal', adjustable='box')
```

```python
# 填充等高线
fig, ax = plt.subplots()

CS = ax.contourf(xx1, xx2, ff, levels=20, cmap='RdYlBu_r')

fig.colorbar(CS)
ax.set_title('$f(x_1, x_2) = x_1 \exp(-x^2_1-x^2_2)$', fontsize=10)
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlim(xx1.min(), xx1.max())
ax.set_ylim(xx2.min(), xx2.max())
ax.grid(False)
ax.set_aspect('equal', adjustable='box')
```
