### 检查常用Python库版本号

#### 查看第三方库版本号

- `Anaconda.Navigator` $\rightarrow$ `Environments`  


- `Anaconda Prompt` $\rightarrow$ `conda list`
- `Anaconda.Navigator` $\rightarrow$ Green triangle $\rightarrow$ `Open Terminal` $\rightarrow$ `pip show numpy`
- `JupyterLab` $\rightarrow$ `!pip list`
- `import numpy`  
  `print(numpy.__version__)`

```python
# !(exclamation mark) is used to run shell or terminal commands in Jupyter Notebook
# !pip list
# !ls
# !pip install numpy
```

```python
# Check the version of commonly used Python libraries
import scipy
print('scipy: %s' % scipy.__version__) # %s: string
```

```python
import numpy
print('numpy: %s' % numpy.__version__)
```

```python
import matplotlib
print('matplotlib: %s' % matplotlib.__version__)
```

```python
import pandas
print('pandas: %s' % pandas.__version__)
```

```python
import statsmodels
print('statsmodels %s' % statsmodels.__version__)
```

```python
import sklearn
print('sklearn: %s' % sklearn.__version__)
```

#### 安装库

- `Anaconda Prompt` $\rightarrow$ `pip install library_name`
- `Anaconda Prompt` $\rightarrow$ `conda install library_name`
- 如果有多个 Anaconda 环境，安装特定库时需要选择特定环境  


- `JupyterLab` $\rightarrow$ `!pip install library_name`
- Download and unpack the source distribution `python setup.py install --user`

#### 卸载库

- `Anaconda Prompt` $\rightarrow$ `pip uninstall library_name`

#### 更新库

- `Anaconda Prompt` $\rightarrow$ `pip install library_name —upgrade`
- `Anaconda.Navigator` $\rightarrow$ `Environments` $\rightarrow$ Blue arrow
