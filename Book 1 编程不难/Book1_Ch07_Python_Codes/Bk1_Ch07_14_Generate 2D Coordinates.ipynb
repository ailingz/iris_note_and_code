### 分离横纵网格坐标

```python
# 自定义函数
# 复刻 numpy.meshgrid()
def custom_meshgrid(x, y):
    num_x = len(x); num_y = len(y)
    X = []; Y = []

    # 外层 for 循环
    for i in range(num_y):
        X_row = []; Y_row = []

        # 内层 for 循环
        for j in range(num_x):
            X_row.append(x[j])
            Y_row.append(y[i])

        # 生成二维数组
        X.append(X_row)
        Y.append(Y_row)

    return X, Y
```

```python
# 示例用法
x = [0, 1, 2, 3, 4, 5] # 横坐标列表
y = [0, 1, 2, 3]       # 纵坐标列表
```

```python
X, Y = custom_meshgrid(x, y)

print(f'X坐标: {X}')
print(f'Y坐标: {Y}')
```
