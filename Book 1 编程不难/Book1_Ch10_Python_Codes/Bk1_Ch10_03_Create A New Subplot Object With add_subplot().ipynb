### 用 `add_subplot()` 方法创建一个新的子图对象

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(211) # 2行1列的第1个位置
# ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(212) # 2行1列的第2个位置

ax1.plot(x, y1)
ax2.plot(x, y2)

plt.show()
```
