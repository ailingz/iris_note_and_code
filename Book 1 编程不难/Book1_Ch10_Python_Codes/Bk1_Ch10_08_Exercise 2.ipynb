```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('_mpl-gallery')

fig, ((ax01, ax02, ax03),
      (ax04, ax05, ax06),
      (ax07, ax08, ax09),
      (ax10, ax11, ax12)) = plt.subplots(4, 3, figsize=(6, 9))

####################################################################################################

# fig 01 - hist()
# make data
np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)

# plot
ax01.hist(x, bins=8, linewidth=0.5, edgecolor='white')

ax01.set(xlim=(0, 8),  xticks=np.arange(1, 8),       xticklabels=[],
         ylim=(0, 56), yticks=np.linspace(0, 56, 9), yticklabels=[])
ax01.set_title('(a) hist()', loc='left', fontsize=9)
ax01.set_aspect(1/7)

####################################################################################################

# fig 02 - boxplot()
# make data
np.random.seed(10)
D = np.random.normal((3, 5, 4),
                     (1.25, 1.00, 1.25),
                     (100, 3)) # random.normal(loc=mean, scale=stdev, size)

# plot
ax02.boxplot(D, positions=[2, 4, 6], widths=1.5, patch_artist=True,
             showmeans=False, showfliers=False,
             medianprops={'color': 'white', 'linewidth': 0.5},
             boxprops={'facecolor': 'C0', 'edgecolor': 'white',
                       'linewidth': 0.5},
             whiskerprops={'color': 'C0', 'linewidth': 0.5},
             capprops={'color': 'C0', 'linewidth': 1.5})

ax02.set(xlim=(0, 8), xticks=np.arange(1, 8), xticklabels=[],
         ylim=(0, 8), yticks=np.arange(1, 8), yticklabels=[])
ax02.set_title('(b) boxplot()', loc='left', fontsize=9)
ax02.set_aspect('equal')

####################################################################################################

# fig 03 - errorbar()
# make data
x = [2, 4, 6]
y = [3.6, 5, 4.2]
yerr = [0.9, 1.2, 0.5]

# plot
ax03.errorbar(x, y, yerr, fmt='o', linewidth=2, capsize=6)

ax03.set(xlim=(0, 8), xticks=np.arange(1, 8), xticklabels=[],
         ylim=(0, 8), yticks=np.arange(1, 8), yticklabels=[])
ax03.set_title('(c) errorbar()', loc='left', fontsize=9)
ax03.set_aspect('equal')

####################################################################################################

# fig 04 - violinplot()
# make data
np.random.seed(10)
D = np.random.normal((3, 5, 4), (0.75, 1.00, 0.75), (200, 3))

# plot
# Create a violin plot which is a method of plotting numeric data
# and can be understood as a combination of a box plot and a kernel density plot
vp = ax04.violinplot(D, [2, 4, 6], widths=2,
                     showmeans=False,
                     showmedians=False,
                     showextrema=False)

# styling
for body in vp['bodies']:
    body.set_alpha(0.5)

ax04.set(xlim=(0, 8), xticks=np.arange(1, 8), xticklabels=[],
         ylim=(0, 8), yticks=np.arange(1, 8), yticklabels=[])
ax04.set_title('(d) violinplot()', loc='left', fontsize=9)
ax04.set_aspect('equal')

####################################################################################################

# fig 05 - eventplot()
# make data
np.random.seed(1)
x = [2, 4, 6]
D = np.random.gamma(4, size=(3, 50))

# plot
ax05.eventplot(D, orientation='vertical', lineoffsets=x, linewidths=0.75)

ax05.set(xlim=(0, 8), xticks=np.arange(1, 8), xticklabels=[],
         ylim=(0, 8), yticks=np.arange(1, 8), yticklabels=[])
ax05.set_title('(e) eventplot()', loc='left', fontsize=9)
ax05.set_aspect('equal')

####################################################################################################

# fig 06 - hist2d()
# make data: correlated + noise
np.random.seed(1)
x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000) / 3

# plot
ax06.hist2d(x, y, bins=(np.arange(-3, 3, 0.1), np.arange(-3, 3, 0.1)))

ax06.set(xlim=(-2, 2), xticks=[], xticklabels=[],
         ylim=(-3, 3), yticks=[], yticklabels=[])
ax06.set_title('(f) hist2d()()', loc='left', fontsize=9)
ax06.set_aspect(2/3)

####################################################################################################

# fig 07 - hexbin()
# make data: correlated + noise
np.random.seed(1)
x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000) / 3

# plot
ax07.hexbin(x, y, gridsize=20)

ax07.set(xlim=(-2, 2), xticks=[], xticklabels=[],
         ylim=(-3, 3), yticks=[], yticklabels=[])
ax07.set_title('(g) hexbin()', loc='left', fontsize=9)
ax07.set_aspect(2/3)

####################################################################################################

# fig 08 - pie()
# make data
x = [1, 2, 3, 4]
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

# plot
ax08.pie(x, colors=colors, radius=3, center=(4, 4),
         wedgeprops={'linewidth': 1, 'edgecolor': 'white'}, frame=True)

ax08.set(xlim=(0, 8), xticks=[], xticklabels=[],
         ylim=(0, 8), yticks=[], yticklabels=[])
ax08.set_title('(h) pie()', loc='left', fontsize=9)
ax08.set_aspect('equal')

####################################################################################################

# fig 09 - tricontour()
# make data
np.random.seed(1)
x = np.random.uniform(-3, 3, 256)
y = np.random.uniform(-3, 3, 256)
z = (1 - x/2 + x**5 + y**3) * np.exp(-x**2 - y**2)
levels = np.linspace(z.min(), z.max(), 15)

# plot
ax09.plot(x, y, 'o', markersize=2, color='lightgrey')
ax09.tricontour(x, y, z, levels=levels)

ax09.set(xlim=(-3, 3), xticks=[], xticklabels=[],
         ylim=(-3, 3), yticks=[], yticklabels=[])
ax09.set_title('(i) tricontour()', loc='left', fontsize=9)
ax09.set_aspect('equal')

####################################################################################################

# fig 10 - tricontourf()
# make data
np.random.seed(1)
x = np.random.uniform(-3, 3, 256)
y = np.random.uniform(-3, 3, 256)
z = (1 - x/2 + x**5 + y**3) * np.exp(-x**2 - y**2)
levels = np.linspace(z.min(), z.max(), 10)

# plot
ax10.plot(x, y, 'o', markersize=1.5, color='grey')
ax10.tricontourf(x, y, z, levels=levels)

ax10.set(xlim=(-3, 3), xticks=[], xticklabels=[],
         ylim=(-3, 3), yticks=[], yticklabels=[])
ax10.set_title('(j) tricontourf()', loc='left', fontsize=9)
ax10.set_aspect('equal')

####################################################################################################

# fig 11 - tripcolor()
# make data
np.random.seed(1)
x = np.random.uniform(-3, 3, 256)
y = np.random.uniform(-3, 3, 256)
z = (1 - x/2 + x**5 + y**3) * np.exp(-x**2 - y**2)

# plot
ax11.plot(x, y, 'o', markersize=1, color='grey')
ax11.tripcolor(x, y, z)

ax11.set(xlim=(-3, 3), xticks=[], xticklabels=[],
         ylim=(-3, 3), yticks=[], yticklabels=[])
ax11.set_title('(k) tripcolor()', loc='left', fontsize=9)
ax11.set_aspect('equal')

####################################################################################################

# fig 12 - triplot()
# make data
np.random.seed(1)
x = np.random.uniform(-3, 3, 256)
y = np.random.uniform(-3, 3, 256)

# plot
ax12.triplot(x, y, linewidth=.5)

ax12.set(xlim=(-3, 3), xticks=[], xticklabels=[],
         ylim=(-3, 3), yticks=[], yticklabels=[])
ax12.set_title('(l) triplot()', loc='left', fontsize=9)
ax12.set_aspect('equal')

####################################################################################################

plt.show()
```
