```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('_mpl-gallery')

fig, ((ax01, ax02, ax03),
      (ax04, ax05, ax06),
      (ax07, ax08, ax09),
      (ax10, ax11, ax12)) = plt.subplots(4, 3, figsize=(6, 9))

####################################################################################################

# fig 01 - plot()
# make data
x = np.linspace(0, 10, 100)
y = 4 + np.sin(2 * x)
x2 = np.linspace(0, 10, 25)
y2 = 4 + np.sin(2 * x2)

# plot
ax01.plot(x2, y2+2.5, 'x', markeredgewidth=2)
ax01.plot(x, y, linewidth=2)
ax01.plot(x2, y2-2.5, 'o-', linewidth=2)

ax01.set(xlim=(0, 8), xticks=np.arange(1, 8), xticklabels=[],
         ylim=(0, 8), yticks=np.arange(1, 8), yticklabels=[])
ax01.set_title('(a) plot()', loc='left', fontsize=9)
ax01.set_aspect('equal')

####################################################################################################

# fig 02 - scatter()
# make data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))

sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
ax02.scatter(x, y, s=sizes, c=colors,
             vmin=0, vmax=100,
             clip_on=False) # clip_on: extends beyond the x/y limits

ax02.set(xlim=(0, 8), xticks=np.arange(1, 8), xticklabels=[],
         ylim=(0, 8), yticks=np.arange(1, 8), yticklabels=[])
ax02.set_title('(b) scatter()', loc='left', fontsize=9)
ax02.set_aspect('equal')

####################################################################################################

# fig 03 - bar()
# make data
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

# plot
ax03.bar(x, y, width=1, edgecolor='white', linewidth=0.7)

ax03.set(xlim=(0, 8), xticks=np.arange(1, 8), xticklabels=[],
         ylim=(0, 8), yticks=np.arange(1, 8), yticklabels=[])
ax03.set_title('(c) bar()', loc='left', fontsize=9)
ax03.set_aspect('equal')

####################################################################################################

# fig 04 - stem()
# make data
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

# plot
ax04.stem(x, y)

ax04.set(xlim=(0, 8), xticks=np.arange(1, 8), xticklabels=[],
         ylim=(0, 8), yticks=np.arange(1, 8), yticklabels=[])
ax04.set_title('(d) stem()', loc='left', fontsize=9)
ax04.set_aspect('equal')

####################################################################################################

# fig 05 - step()
# make data
x = np.arange(8) / 7 * 8
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

# plot
ax05.step(x, y, linewidth=2.5) # plt.stairs(y)

ax05.set(xlim=(0, 8), xticks=np.arange(1, 8), xticklabels=[],
         ylim=(0, 8), yticks=np.arange(1, 8), yticklabels=[])
ax05.set_title('(e) step()', loc='left', fontsize=9)
ax05.set_aspect('equal')

####################################################################################################

# fig 06 - fill_between()
# make data
np.random.seed(1)
x = np.linspace(0, 8, 16)
y1 = 3 + 4 * x/8 + np.random.uniform(0.0, 0.5, len(x))
y2 = 1 + 2 * x/8 + np.random.uniform(0.0, 0.5, len(x))

# plot
ax06.fill_between(x, y1, y2, alpha=.5, linewidth=0)
ax06.plot(x, (y1+y2)/2, linewidth=2)

ax06.set(xlim=(0, 8), xticks=np.arange(1, 8), xticklabels=[],
         ylim=(0, 8), yticks=np.arange(1, 8), yticklabels=[])
ax06.set_title('(f) fill_between()', loc='left', fontsize=9)
ax06.set_aspect('equal')

####################################################################################################

# fig 07 - stackplot()
# make data
x = np.arange(0, 10, 2)
ay = [1, 1.25, 2, 2.75, 3]
by = [1, 1, 1, 1, 1]
cy = [2, 1, 2, 1, 2]
y = np.vstack([ay, by, cy])

# plot
ax07.stackplot(x, y)

ax07.set(xlim=(0, 8), xticks=np.arange(1, 8), xticklabels=[],
         ylim=(0, 8), yticks=np.arange(1, 8), yticklabels=[])
ax07.set_title('(g) stackplot()', loc='left', fontsize=9)
ax07.set_aspect('equal')

####################################################################################################

# fig 08 - imshow()
# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 16), np.linspace(-3, 3, 16))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

# plot
ax08.imshow(Z)

ax08.set(xticks=[], xticklabels=[],
         yticks=[], yticklabels=[])
ax08.set_title('(h) imshow()', loc='left', fontsize=9)
ax08.set_aspect('equal')

####################################################################################################

# fig 09 - contour()
# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
levels = np.linspace(np.min(Z), np.max(Z), 15)

# plot
ax09.contour(X, Y, Z, levels=levels)

ax09.set(xticks=[], xticklabels=[],
         yticks=[], yticklabels=[])
ax09.set_title('(i) contour()', loc='left', fontsize=9)
ax09.set_aspect('equal')

####################################################################################################

# fig 10 - contourf()
# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
levels = np.linspace(Z.min(), Z.max(), 15)

# plot
ax10.contourf(X, Y, Z, levels=levels)

ax10.set(xticks=[], xticklabels=[],
         yticks=[], yticklabels=[])
ax10.set_title('(j) contourf()', loc='left', fontsize=9)
ax10.set_aspect('equal')

####################################################################################################

# fig 11 - quiver()
# make data
x = np.linspace(-4, 4, 6)
y = np.linspace(-4, 4, 6)
X, Y = np.meshgrid(x, y)
U = X + Y
V = Y - X

# plot
ax11.quiver(X, Y, U, V, color='C0', angles='xy',
            scale_units='xy', scale=5, width=.015)
# X, Y: coordinates of the arrow locations
# U, V: components of the vectors (arrows) in the x and y directions, respectively
# angles='xy': indicates that the angle of the arrows should be based on the U and V components,
#              with respect to the Cartesian coordinate system (x and y axes)
# scale_units='xy': specifies that the units of the scaling factor are in the same units as X and Y
#                   This means the length of the arrows will be consistent with the coordinate system
# scale=5: controls the length of the arrows. Specifically, the arrow length will be the length of the vector divided by scale

ax11.set(xlim=(-5, 5), xticks=[], xticklabels=[],
         ylim=(-5, 5), yticks=[], yticklabels=[])
ax11.set_title('(k) quiver()', loc='left', fontsize=9)
ax11.set_aspect('equal')

####################################################################################################

# fig 12 - streamplot()
# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

V = np.diff(Z[1:, :], axis=1)
U = -np.diff(Z[:, 1:], axis=0)

# plot
ax12.streamplot(X[1:, 1:], Y[1:, 1:], U, V) # visualizes the flow of a vector field
# X and Y: the 2D arrays of the x and y coordinates for the grid points
# U and V: the 2D arrays of the x and y components of the vectors at each grid point.
#          Represent the direction and magnitude of the vector field

ax12.set(xticks=[], xticklabels=[],
         yticks=[], yticklabels=[])
ax12.set_title('(l) streamplot()', loc='left', fontsize=9)
ax12.set_aspect('equal')

####################################################################################################

plt.show()
```
